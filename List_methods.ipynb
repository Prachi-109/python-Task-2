# 1. append() – 
**Adds an element at the end of the list**

```python
fruits = ["apple", "banana", "mango"]
fruits.append("orange")
print(fruits)
```

```python
list2 = ["apple","banana"]
list2.append("mango")
print(list2)
```

```python
list3 = [10,20]
list3.append(30)
print(list3)
```

```python
list5 = ["pen","pencil"]
list5.append("eraser")
print(list5)
```

```python
list4 = ["red","blue"]
list4.append("green")
print(list4)
```

# 2. extend() –
**Adds elements of another list**

```python
numbers = [1, 2, 3]
numbers.extend([4, 5, 6])
print(numbers)
```

```python
a = [1,2]
a.extend([3,4])
print(a)
```

```python
e = ["tea"]
e.extend(["coffee","juice"])
print(e)
```

```python
c = [10]
c.extend([20,30])
print(c)
```

```python
d = ["math"]
d.extend(["science","english"])
print(d)
```

# 6. clear() – 
**Removes all elements from the list**

```python
c1 = ['Paper','Book','Bread','Pen','Milk','Apple','Eat']
c1.clear()
print(c1)
```

```python
items = ["pen", "pencil", "eraser"]
items.clear()
print(items)
```

```python
c2 = ['a','b','c','v','a',['r','b','f'],'r','a']
c2.clear()
print(c2)
print(type(c2))
```

# 7. index() – 
**Finds index of an element**

```python
subjects = ["math", "science", "english", "history"]
print(subjects.index("english"))
```

```python
i4 = ['Basic','Dev','Task','Readme','Index','value']
print(f"Index of 'Readme': {i4.index('Readme')}")
```

```python
i2 = [3,5,6,34,5,6,76,7,2,5,4,56]
print(f"Index of 4: {i2.index(4)}")
```

```python
a = [10,20,30]
print(a.index(20))
```

```python
e = ["pen","pencil","eraser"]
print(e.index("pen"))
```

# 8. count() – 
**Counts number of occurrences**

```python
i2 = [3,5,6,34,5,6,76,7,2,5,4,56]
s = i2.count(7)
print(f"No. of times 5 came in list is : {s}")
```

```python
i4 = ['Basic','Dev','Task','Readme','Index','value']
s = i4.count('Task')
print(f"No. of times 'a' came in list is : {s}")
```

```python
numbers = [2, 4, 6, 2, 8, 2]
print(numbers.count(2))
```

```python
a2 = ["apple","banana","apple"]
print(a2.count("apple"))
```

```python
i3 = ['Paper','Pen','Bread','Pen','Milk','Pendrive','Eat']
s = i3.count('Pen')
print(f"No. of times 'a' came in list is : {s}")
```

# 10. reverse() – 
**Reverses the list**

```python
letters = ["a", "b", "c", "d"]
letters.reverse()
print(letters)
```

```python
b = ["apple","banana","mango"]
b.reverse()
print(b)
```

```python
i3 = ['Sayali','Monali','Chakuli','Chetana','Chakuli']
i3.reverse()
print(f"Reverse list is : {i3}")
```

```python
i1 = [3,5,6,34,5,6,76,7,2,5,4,56]
i1.reverse()
print(f"Reverse list is : {i1}")
```

```python
i2 = ['a','r','t','e','a','g','t','s','e','t','a']
i2.reverse()
print(f"Reverse list is : {i2}")
```
